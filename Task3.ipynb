```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find employees with highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

query = "Buy mobile phone buy phone online"

words = query.split()

keyword_count = {}

for word in words:
    if word in keyword_count:
        keyword_count[word] += 1
    else:
        keyword_count[word] = 1

# Display words searched more than once
result = {}

for word, count in keyword_count.items():
    if count > 1:
        result[word] = count

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour in range(len(sensor_readings)):
    if sensor_readings[hour] % 2 == 0:  # even check
        valid_readings.append((hour, sensor_readings[hour]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total = len(emails)

for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
average = sum(sales) / len(sales)

# 30% above average
limit = average * 1.3

for day in range(len(sales)):
    if sales[day] > limit:
        print(f"Day {day + 1}: {sales[day]}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_ids = {}

for user in user_ids:
    if user in count_ids:
        count_ids[user] += 1
    else:
        count_ids[user] = 1

for user, count in count_ids.items():
    if count > 1:
        print(f"{user} → {count} times")
```
