```python
#1. User Login Check
username=input("enter the username")
password=input("enter the password")
if username=='admin' and password=='1234': #checking username and password matches or not
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0
for i in marks:
    if i>=50:
        pass_count=pass_count+1
    else:
        fail_count=fail_count+1
print("total number of pass students",pass_count)
print("total number of fail students",fail_count)
```

```python
#3. Simple Data Cleaner
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]
for i in names:
    cleaned_name=i.strip().lower() #strip:removes extra spaces ---lower:converts to lowercase
    cleaned_names.append(cleaned_name)
print(cleaned_names)
```

```python
# 4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    length=len(i)
    print(length)
    if length>10:
        print("message longer than 10 characters")
```

```python
#5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
    if i=="ERROR":
        count=count+1
print("total error count",count)
```
