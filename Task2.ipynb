```python
#Problem 1: Unique Words in a Sentence
sentence=input("enter a sentence")
def unique_words(sentence):
    words=sentence.split()
    unique_words=set(words)
    print("count of unique words", len(unique_words))
    print("unique_words",unique_words)
unique_words(sentence)
```

```python
#Problem 2: Highest Salary from Employee Data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

def highest_salary(employees):
    for key,value in employees.items():
        if value==max(employees.values()):
            print(f"{key}-{value}")
highest_salary(employees)
```

```python
#Problem 3: Find Maximum and Minimum Values
numbers = [45, 22, 89, 10, 66]
# Initialize max and min with the first element
maximum = numbers[0]
minimum = numbers[0]
# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = [450, 1200, 899, 1500, 300]
count=0
for price in prices:
    if price>1000:
        count=count+1
print("Products above 1000:",count)
```

```python
#Problem 5: Calculate Attendance Percentage
attendance = ["P", "P", "A", "P", "P"]
count=0
for i in attendance:
    if i=='P':
        count=count+1
attendance_percentage=(count/len(attendance))*100
print("Count the number of P values:",count)
print("Attendance Percentage:",attendance_percentage)

```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers=set(phone_numbers)
print("Unique phone numbers:",unique_numbers)
```

```python
#Problem 7: Count Character Frequency
text="ppython"
char_count={}
for char in text:
    if char in char_count:
        char_count[char]+=1
    else:
        char_count[char]=1
print(char_count)
```

```python
#Problem 10: Calculate Average Marks
marks=[34,67,98,78]

def avg_marks(marks):
    total = sum(marks)
    avg = total / len(marks)
    print("Average Marks:", avg)

avg_marks(marks)
```
