```python
#Problem 1: Social Media – Post Engagement Analyzer
def Post_Engagement_Analyzer(likes_list):
    total_likes=0
    for likes in likes_list:
        total_likes+=likes
    if total_likes>=1000:
        status="viral post"
    else:
        status=" Normal Engagement"
    print("Total Likes",total_likes)
    print("Post Status:",status)
likes = [250, 300, 200, 150, 150]
Post_Engagement_Analyzer(likes) 
  
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_adequacy_checker(rainfall_list, required_level):
    
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain
    
    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)
    
    # Conditional check
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    # Output
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)


# Example Usage
rainfall_data = [70, 75, 68, 72, 75]   # Average = 72
rainfall_adequacy_checker(rainfall_data, 70)
```

```python
#Problem 4: Social Media – Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example Usage
users = ["john123", "emma45", "john123", "alex99"]
detect_duplicate_accounts(users)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter
def premium_crop_filter(price_list):
    
    premium_crops = []
    
    # Loop to filter prices
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)


# Example Usage
crop_prices = [1500, 2500, 1800, 3200, 2000]
premium_crop_filter(crop_prices)
```

```python
#Problem 7: System Monitoring – Application Health Checker
def Application_Health_Checker(Errors):
    if Errors == 0:
        status="Healthy"
    elif Errors <= 5:
         status="Minor Issues"
    else:
        status="Critical Issues"
    print("Error Count:",Errors)
    print("System Status:",status)
Application_Health_Checker(7)
```

```python
def attendance_eligibility(attendance_list):
    
    total_classes = 0
    present_count = 0
    
    for status in attendance_list:
        total_classes += 1
        if status == "P":
            present_count += 1
    
    attendance_percentage = (present_count / total_classes) * 100
    
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", eligibility)


# Example Usage
attendance = ["P", "P", "A", "P", "P"]
attendance_eligibility(attendance)
```
