```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount: ₹"))

            # Rule 1: Minimum amount check
            if amount < 50:
                print("Recharge amount must be at least ₹50.")
            
            # Rule 2: Valid plan check
            elif amount not in valid_plans:
                print("Invalid plan selected.")
                print("Available plans:", valid_plans)
            
            else:
                print("Recharge successful for ₹", amount)
                break   # Exit loop if valid

        except ValueError:
            print("Please enter a valid numeric amount.")

# Function Call
validate_recharge()
```

```python
#Problem 2: Inventory Reorder Alert System
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f" Reorder Alert: {product} stock is low ({quantity} units left)")
        else:
            print(f" Stock OK: {product} has sufficient stock ({quantity} units)")

# Sample Dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 18
}

# Function Call
check_inventory(inventory)
```

```python
#Problem 3: Student Result Processing System
def process_result(marks):
    total = 0
    
    # Loop to calculate total marks
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example Usage
student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print("Result:", result)
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if is_peak:
        fare += fare * 0.25   # 25% extra
    
    return fare


# Retry Option
while True:
    try:
        km = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()
        
        peak_status = True if peak == "yes" else False
        
        total_fare = calculate_fare(km, peak_status)
        print("Total Fare: ₹", total_fare)
        
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            break
    
    except ValueError:
        print("Invalid input. Please enter numeric value.")
```

```python
#
def check_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage, "%")
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_eligibility(attendance)
print("Status:", result)
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain digit and @#$)"


# Example
pwd = input("Enter password: ")
print(check_password(pwd))
```
